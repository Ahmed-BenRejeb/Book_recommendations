```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(),cbar=False,ax=ax)
plt.xlabel("columns")
plt.ylabel("missing values")
plt.show()
```

```python
import numpy as np
books["missing_description"] = np.where(books.description.isna(), 1, 0)
books["age_of_book"] = 2025 - books["published_year"]
```

```python
columns_of_interest = ["num_pages","age_of_book","missing_description","average_rating"]

correlation_matrix = books[columns_of_interest].corr(method="spearman")
sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm",cbar_kws={"label": "Spearman correlation"})
heatmap.set_title("Correlation heatmap")
plt.show()
```

```python
book_missing = books[~(books["description"].isna()) & ~(books["published_year"].isna()) & ~ (books["average_rating"].isna()) & ~(books["num_pages"].isna())]
```

```python
book_missing
```

```python
book_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
book_missing["words_in_description"] = book_missing["description"].str.split().str.len()
```

```python
book_missing
```

```python
book_missing.loc[book_missing["words_in_description"].between(1,4), "description"]
```

```python
book_missing_25_word = book_missing[book_missing["words_in_description"]>=25]
```

```python
book_missing_25_word
```

```python
book_missing_25_word["title_and_subtitle"] = (
    np.where(book_missing_25_word["subtitle"].isna(), book_missing_25_word["title"],
    book_missing_25_word[["title","subtitle"]].astype(str).agg(": ".join,axis=1)
))
```

```python
book_missing_25_word
```

```python
book_missing_25_word = book_missing_25_word.drop("titles_and_subtitles", axis=1)
```

```python
book_missing_25_word["tagged_description"] = book_missing_25_word[["isbn13","description"]].astype(str).agg(" ".join,axis =1)
```

```python
book_missing_25_word
```

```python
(
    book_missing_25_word.drop([
        "subtitle","missing_description","age_of_book","words_in_description"],axis=1).to_csv("books_cleaned.csv",index =False)

)
```
